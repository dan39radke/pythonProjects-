#### 2. Dictionary Operations
Perform the following operations on the dictionary called "inventory".

a.  Add the key value pair "stylus": 16      
b.  Update the ipad inventory to 19  
c.  Remove the key value pair, "monitor": 5  
d.  Extract each of the values in the inventory dictionary and add them to the list called counts  

````python
inventory = {
    "iphone": 40,
    "ipad": 17,
    "macbook": 12,
    "monitor": 5    
}

# a. Add the key value pair "stylus": 16 
inventory['stylus'] = 16

# b. Update the ipad inventory to 19
inventory['ipad'] = 19

# c. Remove the key value pair, "monitor": 5
del inventory['monitor']

# d. Extract each of the values in the inventory dictionary and add them to the list called counts
counts = []
for i in inventory.values(): counts = counts + [i]
````

#### Please perform the following

```
print(root.find('Book[@Weight="1.5"]/Authors/Author/Last_Name').text)
```

Using the above as a reference, print the last name of the author who's residence is Beijing  

````python
from lxml import etree
root=etree.XML(data_string)

# a. print all the element tags in the tree
for elements in root.iter(): print(elements.tag)
    

# b. print all the child elements in the tree
for child in root: print(child.tag)
    
# c. print all the author's last names
for elements in root.iter('Author'): print(elements.find('Last_Name').text)
    
# d. The following will print the last name of the book that weighs 1.5 units

print(root.find('Book[@Weight="1.5"]/Authors/Author/Last_Name').text)

# Using the above as a reference, print the last name of the author who's residence is Beijing

print(root.find('Book/Authors/Author/[@Residence="Beijing"]/Last_Name').text)

````

#### 8. Scraping HTML 
In this problem we use the requests library in combination with the BeautifulSoup library to scrape data from the website: https://money.cnn.com/data/markets/.  Take a look at the website to get a sense for its content.  

````python
import requests
from bs4 import BeautifulSoup
url="https://money.cnn.com/data/markets/"
response = requests.get(url)
response.content
results_page=BeautifulSoup(response.content, 'lxml')
print(results_page.prettify())
````

The program below uses the .find() function from BeautifulSoup to extract the text from an html element, <span> with a specific class, "stock-price".

````python
price=results_page.find('span',{'class': 'stock-price'}).get_text()
print(price, type(price))
price=float(price)
print(price, type(price))
````
